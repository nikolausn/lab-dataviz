```python
# setting for showing graphic inline in the notebook
%matplotlib inline
import h5py;
import matplotlib;
import numpy;
# data initialization
data = {};
# read file
with h5py.File("gaia_validp.h5") as dataSet:
    for value in dataSet:
        # read data set
        data[value] = dataSet[value][:];
```

```python
# print keys
# print(data.keys());
# dict_keys(['dec', 'phot_g_mean_mag', 'parallax', 'ra', 'pmra', 'pmdec'])
print(len(data['dec']));
```
